```python
import pandas as pd
```

```python
df=pd.read_csv("salaries.csv")
df
```

```python
input=df.drop('salary_more_then_100k',axis='columns')
input
```

```python
target=df['salary_more_then_100k']
target
```

##sincce ML doesnt know to work with letters or strings or chars we have to convert them using labelencoder

##since you have 3 cols,you have to make 3 objects

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
input['company_new']=le_company.fit_transform(input['company'])
input['job_new']=le_job.fit_transform(input['job'])
input['degree_new']=le_degree.fit_transform(input['degree'])
```

#we have labelencoded the 3 columns

```python
input
```

##now drop the columns with letters

```python
input_new=input.drop(['company','job','degree'],axis='columns')
input_new
```

##as you can see the columns hae been dropped

##now train our model
##for decision tree you have to import tree

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

##now same as before,train the model

##you can also train_test_split method(preferable)

```python
model.fit(input_new,target)
```

```python
model.score(input_new,target)
```

```python
##score is 1 because we are using same dataset to get the output as the dataset we used to train the model
```

```python
model.predict([[2,2,1]])

##means the sales exec working at google has salary <100k
```
